Threshold : 0.8

```python
#Threshold : 0.8
import numpy as np
corr_matrix=df.corr().abs()

#choose from the upper triangle of the matrix
upper=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))

#set threshold as 0.8
to_drop=[column for column in upper.columns if any(upper[column]>0.8)]
print(to_drop)
df_80=df.drop(to_drop,axis=1)

#correlation map
f,ax=plt.subplots(figsize=(18,18))
sns.heatmap(df_80.corr(),annot=True,linewidths=.5,fmt='.2f',ax=ax)
plt.show()
```
