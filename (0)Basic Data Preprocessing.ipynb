```python
batters_all = pd.read_csv('/content/drive/MyDrive/aml_NoneOfUsAreBaseballFans/1871_2023.csv', low_memory=False)
```

```python
batters_all.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
batters_all
```

```python
batters_all = batters_all.sort_values(['IDfg', 'Season'])
```

## 2) 1962년 이전 데이터 삭제

```python
before_1862 = batters_all[batters_all['Season'] < 1962]['IDfg'].unique() # 1962년 이전 데이터가 있는 선수들의 'IDfg' 추출

after_1962 = batters_all[~batters_all['IDfg'].isin(before_1862)].copy() # 추출한 'IDfg'로 1962년 이전 데이터가 있는 선수들 삭제
```

```python
after_1962
```

## 3) 연도(Season)가 연속적이지 않은 player 제거

```python
after_1962['is_continuous'] = (after_1962.groupby('IDfg')['Season'].diff().fillna(1) == 1)
after_1962
```

```python
after_1962 = after_1962[after_1962.groupby('IDfg')['is_continuous'].transform('all')]
after_1962
```

```python
after_1962 = after_1962.drop('is_continuous', axis=1)
```

## 4) 연도(Season) 총합이 10년 안 되는 player 제거

```python
batters_10years = after_1962.groupby('IDfg').filter(lambda x: len(x) >= 10)
```

```python
batters_10years
```

## 5) 11년 이상 뛴 선수: 앞 10년, 뒤 10년 더블링

```python
player_career_length = batters_10years.groupby('IDfg')['Season'].count()
```

### 5-1) 딱 10년만 뛴 선수들

```python
batters_just_10years_index = player_career_length[player_career_length == 10].index # 딱 10년만 뛴 선수들

data = []
for player_id in batters_just_10years_index:
    player_data = batters_10years[batters_10years['IDfg'] == player_id]
    data.append(player_data)

batters_just_10years = pd.concat(data, ignore_index=True)

batters_just_10years
```

```python
print(batters_just_10years['IDfg'].nunique())
```

### 5-2) 11년 이상 뛴 선수들

```python
batters_10years_more_index = player_career_length[player_career_length > 10].index # 11년 이상 뛴 선수들

split_data = []

for player_id in batters_10years_more_index:
    player_data = batters_10years[batters_10years['IDfg'] == player_id]

    # First 10 years
    part1_data = player_data.head(10).copy() # 앞 10년 데이터 추출
    part1_data['IDfg'] = f'{player_id}(1)' # 선수 아이디 변경: 아이디(1)
    split_data.append(part1_data)

    # Last 10 years
    part2_data = player_data.tail(10).copy() # 뒤 10년 데이터 추출
    part2_data['IDfg'] = f'{player_id}(2)' # 선수 아이디 변경: 아이디(2)
    split_data.append(part2_data)

batters_split = pd.concat(split_data, ignore_index=True) # 더블링 된 데이터로 데이터프레임 생성

batters_split
```

```python
print(batters_split['IDfg'].nunique())
```

### 10년만 뛴 선수 + 11년 이상 뛴 선수 합치기

```python
batters_10years_total = pd.concat([batters_just_10years, batters_split]).sort_values(["IDfg", "Season"])
```
