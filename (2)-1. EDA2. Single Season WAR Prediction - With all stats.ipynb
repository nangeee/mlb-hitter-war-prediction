# 특성 선택

```python
x = df[['Age', 'G', 'AB', 'PA', 'H', '1B', '2B', '3B', 'HR', 'R', 'RBI', 'BB',
       'IBB', 'SO', 'HBP', 'SF', 'SH', 'GDP', 'SB', 'CS', 'AVG', 'BB%', 'K%',
       'BB/K', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'wOBA', 'wRAA', 'wRC',
       'Bat', 'Rep', 'Pos', 'RAR', 'Spd', 'BsR', 'Def', 'wSB',
       'Off', 'Lg', 'Events']]
y = df['WAR']
```

```python
# NaN 값을 0으로 대체
x.fillna(0, inplace=True)

# Clip very large or very small values
x = x.clip(lower=-1e10, upper=1e10)
```

```python
# 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#훈련세트 테스트세트 나누기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
print("훈련 데이터 shape: ", x_train.shape, y_train.shape)
print("테스트 데이터 shape: ", x_test.shape, y_test.shape)
```

(1) - 1 SelectKBest 를 이용한 특성 선택

```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression

# SelectKBest 객체 생성
select_feature = SelectKBest(score_func=mutual_info_regression, k=10)

# SelectKBest 객체를 특성과 타겟과 함께 fit
select_feature.fit(x_train, y_train)

# 선택된 특성의 인덱스 확인
selected_indices = select_feature.get_support(indices=True)

# 실제 특성 이름 가져오기
selected_feature_names = df.columns[selected_indices]

# 점수와 해당하는 특성 이름 표시
print('점수 목록:', select_feature.scores_)
print('특성 목록:', selected_feature_names)

# 특성 선택 수행
featured_x = select_feature.transform(x_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# 랜덤 포레스트 모델 생성
random_forest_model = RandomForestRegressor(n_estimators=100, random_state=42)

# 특성 선택을 통해 얻은 특성으로 훈련 데이터와 레이블을 사용하여 모델 학습
random_forest_model.fit(featured_x, y_train)

# 테스트 데이터에 대한 특성 선택 수행
x_test_selected = select_feature.transform(x_test)

# 학습된 모델을 사용하여 예측
y_pred = random_forest_model.predict(x_test_selected)

# R-squared 값 계산
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')

# MAE (Mean Absolute Error) 계산
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
```
