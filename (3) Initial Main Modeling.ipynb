```python
data = '/content/drive/My Drive/Colab Notebooks/ALL_mlb_전처리완료-2020년 제거.csv'
```

```python
import pandas as pd
df = pd.read_csv(data)
```

```python
import numpy as np
```

연도별로 column 이름 확인

```python
features = df.columns
print(features[:2])
print(features[2:42])
print(features[42:82])
print(features[82:122])
print(features[122:162])
print(features[162:202])
print(features[202:242])
print(features[-1])
```

independent variable, dependent variable 나누기

```python
x = df[features[1:-1]]
y = df['Next_WAR']
```

```python
x_feature = x.columns
print(x_feature)
```

```python
# NaN 값을 0으로 대체
x.fillna(0, inplace=True)

# Clip very large or very small values
x = x.clip(lower=-1e10, upper=1e10)
```

```python
# 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#훈련세트 테스트세트 나누기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
print("훈련 데이터 shape: ", x_train.shape, y_train.shape)
print("테스트 데이터 shape: ", x_test.shape, y_test.shape)
```

# **SelectKBest** 를 이용한 특성 선택. (10개)

```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression

# SelectKBest 객체 생성
select_feature = SelectKBest(score_func=mutual_info_regression, k=10)

# SelectKBest 객체를 특성과 타겟과 함께 fit
select_feature.fit(x_train, y_train)

# 선택된 특성의 인덱스 확인
selected_indices = select_feature.get_support(indices=True)

# 실제 특성 이름 가져오기
selected_feature_names = df.columns[selected_indices]

# 점수와 해당하는 특성 이름 표시
print('점수 목록:', select_feature.scores_)
print('특성 목록:', selected_feature_names)

# 특성 선택 수행
featured_x = select_feature.transform(x_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# 랜덤 포레스트 모델 생성
random_forest_model = RandomForestRegressor(n_estimators=100, random_state=42)

# 특성 선택을 통해 얻은 특성으로 훈련 데이터와 레이블을 사용하여 모델 학습
random_forest_model.fit(featured_x, y_train)

# 테스트 데이터에 대한 특성 선택 수행
x_test_selected = select_feature.transform(x_test)

# 학습된 모델을 사용하여 예측
y_pred = random_forest_model.predict(x_test_selected)

# R-squared 값 계산
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')

# MAE (Mean Absolute Error) 계산
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
```

# **RFECV** 이용한 특성선택

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFECV
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split

# RFECV를 사용하여 특성 선택
rfr = RandomForestRegressor()
rfecv = RFECV(estimator=rfr, step=1, cv=5, scoring='r2')
rfecv = rfecv.fit(x_train, y_train)
```

```python
print('Optimal number of features :', rfecv.n_features_)
print('Best features :', x.columns[rfecv.support_])

# 선택된 특성으로 데이터 변환
x_train_selected = rfecv.transform(x_train)
x_test_selected = rfecv.transform(x_test)

# 랜덤 포레스트 회귀 모델 학습
rf_reg = RandomForestRegressor()
rf_reg.fit(x_train_selected, y_train)

# 테스트 데이터에 대한 예측
y_pred = rf_reg.predict(x_test_selected)

# R-squared 값 계산
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')

# MAE (Mean Absolute Error) 계산
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
```

# **PCA**

```python
#주성분 분석
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
```

```python
#주성분 분석
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(x_train)

# 설명 분산 확인 및 누적 설명 분산 계산
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# 최적 주성분 수 찾기 (누적 설명 분산이 95% 이상이 되는 지점)
optimal_n_components = np.argmax(cumulative_explained_variance >= 0.80) + 1

# 결과 출력
print("누적 설명 분산: ", cumulative_explained_variance)
print("최적 주성분 수: ", optimal_n_components)
```

```python
from sklearn.decomposition import PCA

# 최적 주성분 수로 PCA 수행
optimal_pca = PCA(n_components=22)
x_train_optimal_pca = optimal_pca.fit_transform(x_train)

# 결과 출력
print("원본 훈련 데이터 shape: ", x_train.shape)
print("최적 주성분 수로 PCA 적용 후 훈련 데이터 shape: ", x_train_optimal_pca.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_absolute_error


# 랜덤 포레스트 모델 생성
rf_model = RandomForestRegressor()

param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='r2',  verbose=2)

grid_search.fit(x_train, y_train)

# 최적의 모델 확인
best_rf_model = grid_search.best_estimator_

# 테스트 세트에 대한 예측
y_pred = best_rf_model.predict(x_test)

# 평가 지표 출력 (예를 들면, 평균 제곱 오차)
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error on Test Set: {mae}')

# 최적의 하이퍼파라미터 출력
print(f'Best Hyperparameters: {grid_search.best_params_}')
```
