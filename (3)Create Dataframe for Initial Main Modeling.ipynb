## 1) 데이터 불러오기 - 10년 이상 뛴 모든 타자

```python
df = pd.read_csv('/content/drive/MyDrive/aml_NoneOfUsAreBaseballFans/mlb_batters_final.csv', low_memory=False)
df
```

## 2) 전처리  
2-1. null값 제거  
2-2. 2020 시즌 데이터 삭제  
2-3. 2020 시즌 제거 후, 총 10년 안 되는 선수 삭제  
2-4. 필요없는 칼럼 삭제  
2-5. Home runs, Runs 데이터 보정  
2-6. RAB 열 생성: 규정타석(100타석) 만족하지 못한 선수들을 구분하기 위한 것

```python
df = df.dropna(axis=1) # null값 있는 열 삭제
```

```python
# 시즌 2020인 row 모두 삭제
df = df.drop(df[df['Season'] == 2020].index)
df
```

```python
# 10년 안 되는 선수 삭제
df = df.groupby('IDfg').filter(lambda x: len(x) >= 10)
df
```

```python
# 필요없는 칼럼 삭제
# 'Season',  'Name','Team','Age Rng', 'Lg', 'Events','L-WAR', 'Rep' 드랍
df2 = df.drop(columns = ['Name','Team','Age Rng', 'Lg', 'Events','L-WAR', 'Rep'], axis=1)
features = df2.columns
print(features)
print(len(features))
```

```python
# Home runs, Runs 데이터 보정
df2['HR'] = df['HR'] / df['AB']
df2['R'] = df['R'] / df['PA']
```

```python
# RAB 열 생성: 추후에 규정타석(100타석) 만족하지 못한 선수들을 필터링하기 위한 용도
df2['RAB'] = df2['PA'].apply(lambda x: 1 if x >= 100 else 0)
df2 # 총 21300 rows (2130명)
```

```python
print(df2.columns)
print(len(df2.columns))
```

## 3) 데이터 전처리 2  
3-1. 10년치 데이터 분리: 앞 6년은 하나의 row에 넣고, 뒤의 4년치 데이터에서는 WAR 합만 추출. 이 WAR 합은 'Next_WAR' 칼럼으로 들어감.  

```python
# 칼럼 순서 조정
columns_order = ['IDfg', 'Age', 'Season', 'PA', 'G', 'AB', 'H', '1B', '2B', '3B', 'HR',
                 'R', 'RBI', 'BB', 'IBB', 'SO', 'HBP', 'SF', 'SH', 'GDP', 'SB', 'CS',
                 'AVG', 'BB%', 'K%', 'BB/K', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'wOBA',
                 'wRAA', 'wRC', 'Bat', 'Pos', 'RAR', 'WAR', 'Spd', 'BsR', 'Def', 'wSB',
                 'Off', 'RAB']

df2 = df2[columns_order]
print(df2.columns)
print(len(df2.columns))
```

```python
# 앞 6년치 데이터 추출 위해 칼럼 생성/조정
column_6_seasons = ['IDfg', 'Age']
for i in range(1, 7):
  for col in df2.columns[2:]:
    column_6_seasons.append(f'{col}_{i}')

column_6_seasons.append('Next_WAR')
print(column_6_seasons)
len(column_6_seasons)
```

```python
# 위에서 생성한 칼럼으로 새로운 데이터 프레임 만들기
df3 = pd.DataFrame(columns=column_6_seasons)
```

```python
# 앞 6년은 하나의 row에 넣고, 뒤의 4년치 데이터에서는 WAR 합만 추출
season = 0
Next_WAR = 0

for index, row in df2.iterrows():  # 10년치 데이터가 모두 담긴 df의 row를 순회
  season += 1

  if season == 1: # 첫 번째 시즌: player_season_data에 'IDfg', 'Age'열 데이터 추가
    player_data = [row['IDfg'], row['Age'] + 6]
    player_season_data = []

  if season <= 6: # 1~6번째 시즌: player_season_data에 'Season', 'PA', 'G', ... 'Off', 'RAB' 값 모두 추가
    for col in df2.columns[2:]:
      player_season_data.append(row[col])

  if season >= 7 and season <= 10: # 7~10번째 시즌: 해당 시즌의 WAR 값 Next_WAR에 더하기
    Next_WAR += row['WAR']

  if season == 10: # 10번째 시즌: player_season_data에 마지막 row 'Next_WAR' 추가 + player_data와 player_season_data로 6년치 데이터가 하나의 row에 담긴 데이터프레임 생성
    player_season_data.append(Next_WAR)
    df3 = pd.concat([df3, pd.DataFrame([player_data + player_season_data], columns=column_6_seasons)], ignore_index=True)
    Next_WAR = 0
    season = 0
```

## 4) 데이터 전처리 3  
4-1. 4시즌 이상 규정타석(100타석) 만족하지 못한 선수들 필터링  
4-2. 규정타석(100타석) 만족하지 못한 시즌의 데이터: 나머지 시즌 데이터의 평균값으로 대체

```python
# 4시즌 이상 규정타석(100) 만족한 선수들만 필터링
df3['RAB']=df3['RAB_1']+df3['RAB_2']+df3['RAB_3']+df3['RAB_4']+df3['RAB_5']+df3['RAB_6']
df3 = df3[df3['RAB'] >= 4]
df3 = df3.reset_index(drop=True)
```

```python
df3 # 필터링 후: 선수 총 1588명
```

```python
cols = ['G', 'AB', 'H', '1B', '2B', '3B', 'HR',
       'R', 'RBI', 'BB', 'IBB', 'SO', 'HBP', 'SF', 'SH', 'GDP', 'SB', 'CS',
       'AVG', 'BB%', 'K%', 'BB/K', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'wOBA',
       'wRAA', 'wRC', 'Bat', 'Pos', 'RAR', 'WAR', 'Spd', 'BsR', 'Def', 'wSB',
       'Off', 'RAB']
RAB_zero_columns = [] # RAB가 0일 때 평균값으로 대체할 칼럼들
for i in range(1, 7):
  season_data = []
  for col in cols:
    season_data.append(f'{col}_{i}')
  RAB_zero_columns.append(season_data)

print(RAB_zero_columns)
```

```python
df4 = df3.copy()
```

```python
for index, row in df4.iterrows():
  sum_6 = 0 # 6시즌 데이터의 총합
  for i in range(6):
    sum_6 += ((row[RAB_zero_columns[i]] * row[f'RAB_{i+1}']).values)
  average = sum_6 / row['RAB'] # 6시즌 데이터의 평균

  for i in range(6):
    if row[f'RAB_{i + 1}'] == 0: # RAB 값이 0인 시즌은
      df4.loc[index, RAB_zero_columns[i]] = average # 나머지 시즌 데이터의 평균값을 대체
```

```python
# 사용 후, 필요없어진 칼럼 제거
df4 =df4.drop(['RAB', 'RAB_1', 'RAB_2', 'RAB_3', 'RAB_4', 'RAB_5', 'RAB_6'],axis=1)
df4 =df4.drop(['Season_1', 'Season_2', 'Season_3', 'Season_4', 'Season_5', 'Season_6'],axis=1)
```
