남아 있는 선수들 중 타석 조건 못 채운 선수들의 데이터 보정
RAB가 0인 해의 데이터 = RAB가 1인 해의 데잍

```python
#RAB 0인 시즌은 RAB=1인 시즌의 평균값으로 대체
year1=range(4,14) #'HR1','R1','AVG1','BB%1','OBP1','SLG1','Pos1','WAR1','BSR1','Def1'
year2=range(17,27) #'HR2','R2','AVG2','BB%2','OBP2','SLG2','Pos2','WAR2','BSR2','Def2'
year3=range(30,40)
year4=range(43,53)
year5=range(56,66)
year6=range(69,79)

#2020년도의 RAB value를 0으로 바꾸기
for i in range(df3.shape[0]):
  for j in range(1,7):
    if df3.loc[i,'Season'+str(j)]==2020:
      df3.loc[i,'RAB'+str(j)]=0

for i in range(df3.shape[0]): #데이터프레임 전체 탐색
  #평균값 계산.
  #i번째 인덱스의 선수가 있을 때,
  #year1~6데이터들과 해당 년도 RAB를 곱함. => 1일 경우 그대로, 0일 경우 전부 다 0으로 바뀜
  #다 더하고, 규정 타석 채운 연도 수로 나누기
  Average=((df3.iloc[i,year1]*df3.loc[i,'RAB'+'1']).values+(df3.iloc[i,year2]*df3.loc[i,'RAB'+'2']).values+(df3.iloc[i,year3]*df3.loc[i,'RAB'+'3']).values+
           (df3.iloc[i,year4]*df3.loc[i,'RAB'+'4']).values+(df3.iloc[i,year5]*df3.loc[i,'RAB'+'5']).values+(df3.iloc[i,year6]*df3.loc[i,'RAB'+'6']).values)/df3.loc[i,'RAB']
  PA_Average=(df3.loc[i,'PA1']+df3.loc[i,'PA2']+df3.loc[i,'PA3']+df3.loc[i,'PA4']+df3.loc[i,'PA5']+df3.loc[i,'PA6'])/6
  replacing_years=[]
  for j in range(1,7): #6년치 RAB 열 확인
    if (df3.loc[i,'RAB'+str(j)]==0): #RAB가 0이면
      #평균값으로 대체
      df3.iloc[i,locals()['year'+str(j)]]=Average
    if df3.loc[i,'Season'+str(j)]==2020:#Season이 2020이면
      df3.loc[i,locals()['PA'+str(j)]]=PA_Average #PA 평균값으로 대체
      df3.iloc[i,locals()['year'+str(j)]]=Average #기록도 평균값으로 대체
```
